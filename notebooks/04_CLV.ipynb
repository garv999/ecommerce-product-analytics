# Customer Lifetime Value (CLV) Analysis
---


```python
import pandas as pd
import numpy as np

df_clean = pd.read_csv(
    "../data/processed/cleaned_online_retail.csv",
    parse_dates=["InvoiceDate"]
)

df_clean.head()
```

```python
customer_revenue = (
    df_clean.groupby("Customer ID")["Revenue"]
    .sum()
    .reset_index()
)

customer_revenue.head()
```

```python
avg_revenue_customer = customer_revenue["Revenue"].mean()

print("Average Revenue per Customer:", round(avg_revenue_customer, 2))
```

```python
purchase_frequency = (
    df_clean.groupby("Customer ID")["Invoice"]
    .nunique()
    .mean()
)

print("Average Purchase Frequency:", round(purchase_frequency, 2))
```

```python
#CLV=Average Revenue per Customer×Purchase Frequency
clv = avg_revenue_customer * purchase_frequency

print("Estimated Customer Lifetime Value:", round(clv, 2))
```

## Revenue Concentration
---

```python
customer_revenue_sorted = customer_revenue.sort_values(
    by="Revenue", ascending=False
)

top_10_percent = int(len(customer_revenue_sorted) * 0.1)

top_revenue = customer_revenue_sorted.head(top_10_percent)["Revenue"].sum()
total_revenue = customer_revenue_sorted["Revenue"].sum()

print(
    "Top 10% customers contribute:",
    round((top_revenue / total_revenue) * 100, 2),
    "% of revenue",
)
```

## Customer Lifetime Value Insights
---
### Customer Revenue Potential
The estimated **Customer Lifetime Value (CLV)** represents the total revenue a business can expect from an average customer over their purchasing lifecycle. Understanding CLV helps businesses determine how much they can invest in acquiring and retaining customers.

### Revenue Concentration
Analysis shows that a **small percentage of customers contribute a disproportionately large share of total revenue**. This highlights the importance of identifying and retaining high-value customers.

### Business Implication
Increasing **customer retention and purchase frequency** can significantly improve CLV, making retention strategies more cost-effective than constantly acquiring new customers.
